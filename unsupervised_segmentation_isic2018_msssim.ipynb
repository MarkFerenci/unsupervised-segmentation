```python
from PIL import Image
from glob import glob
import pytorch_msssim
import torch
import time
import numpy as np
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def load_images(files: list):
  ims = []
  for f in files:
    ims.append(np.asarray(Image.open(f)))
  return np.asarray(ims)
```

```python
def img_to_torch(img):
  r, g, b = img[:,:,0], img[:,:,1], img[:,:,2]
  return np.array([r, g, b])
```

```python
start = time.time()
msssim_scores = []

for epoch in range(10, 101, 10):
  sum = 0
  print(f'epoch: {epoch}')

  files = glob(f'pix2ultra/results/pix2ultra/test_{epoch}/images/*_synthesized_image.jpg')
  synth = load_images(files)
  files = glob('isic-train-100/*.jpg')
  orig = load_images(files)
  synth, orig = synth[:10], orig[:10]
  print('Loaded', synth.shape, orig.shape)
  synth = synth.astype('float32')
  orig = orig.astype('float32')
  
  for i, s in enumerate(synth):
    s = img_to_torch(s)
    s = s[np.newaxis, ...]
    s = torch.from_numpy(s)
    for j, o in enumerate(orig):
      print(f'compare synthetic image {i} with original image {j}')
      o = img_to_torch(o)
      o = o[np.newaxis, ...]
      o = torch.from_numpy(o)
      score = pytorch_msssim.msssim(s, o).item()
      sum += score
  avg_msssim = sum/len(synth)/len(orig)
  print(f'average MS-SSIM: {avg_msssim:.3f}')
  msssim_scores.append((epoch, avg_msssim))

msssim_scores = sorted(msssim_scores, key=lambda s: s[1])
best_epoch, best_msssim = msssim_scores[-1]

print(f'Best epoch is {best_epoch} with MS-SSIM: {best_msssim}')
print('All values')
for e, s in msssim_scores:
  print(f'For epoch {e} MS-SSIM is {s:.3f}')

end = time.time()
print(f'Duration: {end-start} s')
```
